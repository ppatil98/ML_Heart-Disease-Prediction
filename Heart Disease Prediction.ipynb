```python
per_25 = data1.serum_cholesterol_mg_per_dl.quantile(0.25)
per_75 = data1.serum_cholesterol_mg_per_dl.quantile(0.75)
iqr = per_75-per_25
upper_limit = per_75+(1.5*iqr)
lower_limit = per_25-(1.5*iqr)
data1.serum_cholesterol_mg_per_dl = np.where(data1['serum_cholesterol_mg_per_dl']>upper_limit,upper_limit,np.where(data1.serum_cholesterol_mg_per_dl<lower_limit,lower_limit,data1.serum_cholesterol_mg_per_dl))
```

```python
def outlier_handle(d):
  per_25 = data1[d].quantile(0.25)
  per_75 = data1[d].quantile(0.75)
  iqr = per_75-per_25
  upper_limit = per_75+(1.5*iqr)
  lower_limit = per_25-(1.5*iqr)
  data1[d] = np.where(data1[d]>upper_limit,upper_limit,np.where(data1[d]<lower_limit,lower_limit,data1[d]))
```

```python
dp =['resting_blood_pressure','chest_pain_type','serum_cholesterol_mg_per_dl','oldpeak_eq_st_depression','num_major_vessels']
```

```python
for i in dp :
  outlier_handle(i)
```

```python
X= data1
Y = data.heart_disease_present
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
smote = SMOTE(random_state=42)
X_smote, Y_smote = smote.fit_resample(X,Y)  #balencing the data
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_smote)        #scaling the data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y_smote,test_size = 0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier #random forest classifier

rf_clf = RandomForestClassifier(n_estimators=100)
rf_clf.fit(X_train,Y_train)
```

```python
y_predict=rf_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score
print(classification_report(Y_test,y_predict))
```

```python
from sklearn.model_selection import cross_val_score, KFold
scores = cross_val_score(rf_clf,X_scaled,Y_smote, cv=4)
```

```python
print(scores)
```

```python
print(scores.mean())
```

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
```

```python
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}
```

```python
rf_clf1 = RandomForestClassifier(random_state=42)
rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='f1',param_distributions=random_grid, n_iter=100, cv=3,
                               verbose=2, random_state=42, n_jobs=-1)
```

```python
rf_cv.fit(X_train, Y_train)
```

```python
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")
```
